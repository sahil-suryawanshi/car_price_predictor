## Cleaning

```python
backup = df.copy()
```

```python
df = df[df["year"].str.isnumeric()]
```

```python
df["year"] = df["year"].astype(int)
```

```python
df.info()
```

```python
df = df[df["Price"]!= "Ask For Price"]
```

```python
df["Price"] = df["Price"].str.replace(",","").astype(int)
```

```python
df.info()
```

```python
df["kms_driven"] = df["kms_driven"].str.split(" ").str.get(0).str.replace(",","")
```

```python
df=df[df["kms_driven"].str.isnumeric()]
```

```python
df.info()
```

```python
df["kms_driven"] = df["kms_driven"].astype(int)
```

```python
df.info()
```

```python
df = df[~df["fuel_type"].isna()]  # ~  = means excluding this row
```

```python
df["name"] = df["name"].str.split(" ").str.slice(0,3).str.join(" ")
# split use for spliting every word by space, with the slice we getting first three words after we join it
```

```python
df = df.reset_index(drop = True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[df["Price"] < 6e6].reset_index(drop = True)  #60lakh means 6e6
```

```python
df
```

```python
df.to_csv("Cleaned car.csv")
```

## Model

```python
X = df.drop(columns = "Price")
y = df["Price"]
```

```python
X.shape
```

```python
y.shape  #Target Column
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[["name","company","fuel_type"]])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_),["name","company","fuel_type"]),
                                       remainder = "passthrough")
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
Scores = []
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    Scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(Scores)
```

```python
Scores[np.argmax(Scores)]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=np.argmax(Scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```
